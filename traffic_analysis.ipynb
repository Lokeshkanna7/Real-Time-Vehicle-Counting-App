```python
import torch
import cv2

# Load pre-trained YOLOv5 model
model = torch.hub.load('ultralytics/yolov5', 'yolov5s')  \

def classify_vehicle(label):
    heavy = ['truck', 'bus']
    auto = ['car', 'motorcycle', 'van']
    if label in heavy:
        return 'heavy'
    elif label in auto:
        return 'auto'
    else:
        return 'other'

cap = cv2.VideoCapture(r"D:\nycdot\large.mp4")
fps = cap.get(cv2.CAP_PROP_FPS)
print(f"Video FPS: {fps}")
frame_count = 0
vehicle_counts = {'auto': 0, 'heavy': 0}

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    frame_count += 1
    if frame_count % 3 != 0:  
        continue

    results = model(frame)
    detections = results.pandas().xyxy[0]

    for _, row in detections.iterrows():
        label = row['name']
        vehicle_type = classify_vehicle(label)
        if vehicle_type in vehicle_counts:
            vehicle_counts[vehicle_type] += 1

    annotated_frame = results.render()[0]
    cv2.imshow('Detection', annotated_frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()

print("Vehicle Counts:", vehicle_counts)
```

```python
#prototype for real-time vehicle detection and not tested yet

import torch
import cv2

model = torch.hub.load('ultralytics/yolov5', 'yolov5s', pretrained=True)
model.eval()

def classify_vehicle(label):
    heavy = ['truck', 'bus']
    auto = ['car', 'motorcycle', 'van']
    if label in heavy:
        return 'heavy'
    elif label in auto:
        return 'auto'
    else:
        return 'other'

cap = cv2.VideoCapture(0)
vehicle_counts = {'auto': 0, 'heavy': 0}
frame_count = 0

while True:
    ret, frame = cap.read()
    if not ret:
        break
    frame_count += 1
    if frame_count % 5 != 0:  
        continue

    results = model(frame)
    detections = results.pandas().xyxy[0]

    for _, row in detections.iterrows():
        label = row['name']
        vehicle_type = classify_vehicle(label)
        if vehicle_type in vehicle_counts:
            vehicle_counts[vehicle_type] += 1

    annotated_frame = results.render()[0]
    cv2.putText(annotated_frame, f"Auto: {vehicle_counts['auto']}  Heavy: {vehicle_counts['heavy']}", 
                (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

    cv2.imshow('Real-Time Vehicle Detection', annotated_frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()

print("Final Vehicle Counts:", vehicle_counts)
```
